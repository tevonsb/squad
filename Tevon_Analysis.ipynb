```python
def average_length(predictions):
    count = 0
    average = 0
    for pred in predictions:
        average += len(predictions[pred])
        count += 1
    return average/count

def average_for_array(predictions):
    count = 0
    average = 0
    for pred in predictions:
        count += 1
        if len(predictions[pred]) == 0:
            continue
        lens = [len(s) for s in predictions[pred]]
        total = sum(lens)/len(lens)
        average += total
    return average/count

def max_length(predictions):
    count = 0
    average = 0
    for pred in predictions:
        count += 1
        if len(predictions[pred]) == 0:
            continue
        lens = [len(s) for s in predictions[pred]]
        total = max(lens)
        average += total
    return average/count

print("Our average answer length is:")
print(average_length(our_predictions))

print("Bert average asnwer length is:")
print(average_length(bert_predictions))

print("Golden answers average length of average length:")
print(average_for_array(dev_predictions))

print("Golden answers average length of max lengths:")
print(max_length(dev_predictions))
```

```python
# Predicting the right <no-answer>
# Returns percent percent of questions where an answer was given incorrectly, percent of questions where
# an answer was not given incorrectly
def correct_no_answers(truth, predictions):
    count_answer_for_no_answer = 0
    count_no_answer_for_answer = 0
    count = 0
    for t in truth:
        count += 1
        pred = predictions[t]
        if len(truth[t]) == 0:
            if pred is not '':
                count_answer_for_no_answer += 1
        if len(truth[t]) is not 0:
            if pred is '':
                count_no_answer_for_answer += 1
    
    return count_answer_for_no_answer/count, count_no_answer_for_answer/count

bert_answer, bert_no_answer = correct_no_answers(dev_predictions, bert_predictions)
our_answer, our_no_answer = correct_no_answers(dev_predictions, our_predictions)
print("Bert answered when it shouldn't have:")
print(bert_answer)
print("Bert did not answer when it should have:")
print(bert_no_answer)

print("We answered when it shouldn't have:")
print(our_answer)
print("We did not answer when it should have:")
print(our_no_answer)
                
```
